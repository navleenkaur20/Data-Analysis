Q5. The gold rates in CNY for the dates 1/1/1985 – 1/4/1985 are missing. 1 USD equals to 2.82 CNY
(Average USD to CNY conversion rate in January 1985). Fill the missing CNY values for these
dates based on the USD rate given at respective dates (10 points)

```python

conversion_rate = 2.82

mask = gold['CNY'].isna()

gold['CNY'] = gold['CNY'].mask(mask, gold['USD'] * conversion_rate)

print(gold.head(5))
```

Q6.Slice the year 2021 dataset from “gold” into new data frame called “gold_rate_2021” and
execute the following: [2+10+3]

1.   List the average, highest, and lowest gold rate for INR by months. Refer the sample
output given below:
2.   Which month of the year the gold rate was lowest in USD, EUR, and INR? Print the
values with month name.





```python
gold_2021 = gold[gold['Date'].str.contains('2021')]
print(gold_2021)
```

```python
import pandas as pd

gold_2021 = gold[pd.to_datetime(gold['Date'], format='%m/%d/%Y').dt.year == 2021].copy()


gold_2021['Date'] = pd.to_datetime(gold_2021['Date'], format='%m/%d/%Y')

gold_2021['Month'] = gold_2021['Date'].dt.month_name()


month_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

gold_2021['Month'] = pd.Categorical(gold_2021['Month'], categories=month_order, ordered=True)


gold_stats_2021 = gold_2021.groupby(['Month'], observed=False)['INR'].agg(['mean', 'max', 'min'])


gold_stats_2021.columns = pd.MultiIndex.from_tuples(
    [('INR', 'mean'), ('INR', 'max'), ('INR', 'min')]
)


gold_stats_2021 = gold_stats_2021.reset_index()


print(gold_stats_2021.sort_values(by=['Month'], key=lambda x: pd.Categorical(x, categories=month_order, ordered=True)))




```

```python
gold_2021.head()
```

```python

lowest_gold_usd = gold_2021.loc[gold_2021['USD'].idxmin()]
lowest_gold_eur = gold_2021.loc[gold_2021['EUR'].idxmin()]
lowest_gold_inr = gold_2021.loc[gold_2021['INR'].idxmin()]


print(f"Lowest gold rate in USD: Month={lowest_gold_usd['Month']}, Value={lowest_gold_usd['USD']}")
print(f"Lowest gold rate in EUR: Month={lowest_gold_eur['Month']}, Value={lowest_gold_eur['EUR']}")
print(f"Lowest gold rate in INR: Month={lowest_gold_inr['Month']}, Value={lowest_gold_inr['INR']}")
```

Q7.Add a new column called “Remarks” that should be filled as “gold rally “or “bearish gold
market” based on gold rates in GBP in the previous date and the next date. Refer the example
given below [10 points].
You must use the shift () and lambda functions to arrive at the solution

```python
gold['Remarks'] = gold.apply(
    lambda row: np.nan if row.name == 0 else
                 ('gold rally' if row['GBP'] > gold['GBP'].shift(1)[row.name] and row['GBP'] < gold['GBP'].shift(-1)[row.name]
                  else 'bearish gold market'),
    axis=1
)


print(gold)
```


1.   2008 Financial Crisis Peak:


*   The significant spike in gold prices around 2008 can be attributed to the global financial crisis. During times of economic instability, investors tend to flock towards gold as a safe-haven asset, leading to a surge in demand and, consequently, higher prices.
*  Reference: During the global recession, central banks and governments around the world took measures to address the financial crisis, including massive fiscal stimulus packages, leading to concerns over inflation and the value of paper currencies. This drove up gold prices as a hedge.

2. 2020 COVID-19 Pandemic Surge:


*   The sharp rise in gold prices around 2020 is tied to the COVID-19 pandemic and the resulting economic uncertainty. As governments worldwide implemented lockdowns and issued economic stimulus packages, concerns over inflation and potential currency devaluation increased, boosting gold's appeal.
*   Reference: In 2020, amid global uncertainty and financial stimulus measures, the price of gold rose sharply. The World Gold Council reported a significant increase in gold demand due to its safe-haven status during the global crisis.
